```python
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
import windrose
#import proplot as pplt
```

```python
# data
data = pd.read_csv("Fig_3_wind_djfma.csv")
```

```python
data.shape
```

```python
import matplotlib 
matplotlib.rc('xtick', labelsize=12)
matplotlib.rc('ytick', labelsize=12)

ax = WindroseAxes.from_ax()
ax.bar(data.WD, data.WS, normed=True, opening=0.8, edgecolor='white', cmap=cm.plasma)
#ax.set_legend(bbox_to_anchor=(0.6,0.5))
plt.legend(title="Wind speed [m/s]", fontsize=10, title_fontsize=10,
          bbox_to_anchor=(0.7,0.515))

#ax.set_xticklabels((90, 45, 0, 315, 270, 225, 180, 135))         # https://github.com/python-windrose/windrose/issues/151
ax.set_xticklabels(['E', 'NE', 'N', 'NW',  'W', 'SW', 'S', 'SE']) # https://stackoverflow.com/questions/65129447/python-windrose-90-degrees-off-and-wrong-labels

plt.suptitle('AWS-M [DJFMA; n = 69666]', size=18)
plt.show()
#plt.savefig('py_windrose_revision_proplot.jpeg', dpi=300)
```
